Creating Pandas DataFrame from numpy array

```python
data = np.array([
    ["Alice",'25','Russia'],
    ["BOb","23","America"],
    ["Aryan","21","India"]
])
df = pd.DataFrame(data,columns=["Name","Age","Country"])
df
```

Creating Pandas DataFrame from Dictionary

```python
data = {
    'Name':['Ram',"Krishan","Aryan"],
    'Age': [np.Inf,np.Inf,21],
    'Place': ['Ayodhya', 'Mathura', 'Orai']
}
df = pd.DataFrame(data)
df
```

Creating DataFrame from List of lists

```python
data = [
    ['RAM',np.Inf,'Ayodhya'],
    ["KRISHAN",np.Inf,"Mathura"],
    ["Aryan",21,"Orai"]
]
df = pd.DataFrame(data,columns=["Name","Age","Place"])
df
```

Creating Pandas DataFrame from Tuples

```python
data = [
    ('RAM',np.Inf,'Ayodhya'),
    ("KRISHAN",np.Inf,"Mathura"),
    ("Aryan",21,"Orai")
]
df = pd.DataFrame(data,columns = ["Name","Age","Place"])
df
```

```python
df = pd.DataFrame([[1,2,3],[4,5,6]],columns=['cl1','cl2','cl3'])
df
```
